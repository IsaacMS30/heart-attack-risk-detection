## Cleaning data

First, we will drop some columns that are not necessary (patient ID).

```python
# Drop unnecessary columns
df = prep.drop_columns(df, ["Patient ID", "Country", "Continent", "Hemisphere"])
df
```

We know there are no missing values. We just need to transform some columns (Sex, Blood Pressure, Diet).

* For **Sex**, we can assign 0 to Female and 1 to Male.

* For **Blood Pressure**, we can split it in two different columns (systolic & diastolic).

* For **Diet**, we can assign 0 to Healthy, 1 to Average and 2 to Unhealthy.

```python
df.dtypes
```

```python
# Transform Sex column
df = prep.transform_sex_col(df, {"Female": 0, "Male": 1})
df
```
